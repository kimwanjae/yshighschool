```python
import threading
import socket
import cv2
import keyboard
import pygame
import sys
import time
from pygame.locals import *
```

```python
tello_address = ('192.168.10.1', 8889)
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
sock.bind(('', 9000))
```

```python
def send(message):
  try:
    sock.sendto(message.encode(), tello_address)
    print("Sending message: " + message)
  except Exception as e:
    print("Error sending: " + str(e))
```

```python
def receive():
  try:
    response, ip_address = sock.recvfrom(128)
    print("Received message: " + response.decode(encoding='utf-8') + " from Tello with IP: " + str(ip_address))
  except Exception as e:
    print("Error receiving: " + str(e))
```

```python
def main():
    

    
    send('command')
    send('streamon')
    time.sleep(3)

    video_dev = cv2.VideoCapture('udp://192.168.10.1:11111')
    video_width = video_dev.get(cv2.CAP_PROP_FRAME_WIDTH)
    video_height = video_dev.get(cv2.CAP_PROP_FRAME_HEIGHT)

    # Main loop
    try:
        prev_timestamp = time.time()

        while True:
            ret, orig_image = video_dev.read()
            curr_time = time.localtime()
            cv2.imshow('Video', orig_image)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

            # Check if the stream is over
            if ret is None or orig_image is None:
                break

            res=''
            if keyboard.is_pressed('t'):
                res='takeoff'
                send(res)
            elif keyboard.is_pressed('up'):
                res='forward 50'
                send(res)
            elif keyboard.is_pressed('left'):
                res='left 50'
                send(res)
            elif keyboard.is_pressed('down'):
                res='back 50'
                send(res)
            elif keyboard.is_pressed('right'):
                res='right 50'
                send(res)
            elif keyboard.is_pressed('a'):
                res='ccw 30'
                send(res)
            elif keyboard.is_pressed('d'):
                res='cw 30'
                send(res)
            elif keyboard.is_pressed('w'):
                res='up 50'
                send(res)
            elif keyboard.is_pressed('s'):
                res='down 50'
                send(res)
            elif keyboard.is_pressed('l'):
                res='land'
                send(res)
                break



    except KeyboardInterrupt:
        print('User interruption')

    # End image device
    send('land')
    time.sleep(5)
    video_dev.release()
    sock.close()


if __name__ == '__main__':
    main()
```
